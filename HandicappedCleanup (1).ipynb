```python
import pandas as pd
```

```python

#school = "   SCHOOL NAME     1234 S. State     HS     Accessible      "
#school_list = school.split("  ")
#fields = [s.strip() for s in school_list if s.strip() != '']
# fields[0] = "SCHOOL NAME"
# fields[1] = "1234 S. State"
# fields[2] = "HS"
# fields[3] = "Accessible"

# Using readlines() 
file1 = open('Handicapdata.txt', 'r')
Lines = file1.readlines() 
count = 0

schoollist=[]
    
# Strips the newline character 
for line in Lines: 
    #line = line.strip()
    lineList = line.split("  ")
    #print(lineList)
    cleanedList = [i.strip() for i in lineList if i]
    #cleanedList = [i for i in lineList if i] = eliminates lines which are empty
    if len(cleanedList)==4:
        schoollist.append(cleanedList)
print(schoollist) 
```

```python
#cleanedList can now be seperated into columns
schoollist_df = pd.DataFrame(schoollist, columns =['School_Name', 'Address', 'School_Type', 'Accessible'])

#cleanedList_df.to_csv('CleanedList.csv', encoding='utf-8', index=False)
print(schoollist_df)
```
